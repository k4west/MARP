```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# sns.distplot(tips['total_bill'], bins=40) # This function has been deprecated and will be removed in seaborn v0.14.0. 

sns.histplot(tips['total_bill'], bins=40, kde=True, stat="density", kde_kws=dict(cut=3))
# sns.histplot(tips['total_bill'], bins=40, kde=True, stat="density")
```

```python
sns.displot(tips['total_bill'], kind='kde')
# sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x ='total_bill', y ='tip', data=tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.countplot(x='day', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
tc = tips.corr(numeric_only=True)
sns.heatmap(tc, annot=True)
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
fp
```

```python
sns.heatmap(fp)
```

```python
sns.clustermap(fp, cmap='coolwarm')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'])
```
